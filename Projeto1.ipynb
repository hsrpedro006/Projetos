*  Renda Anual vs. Score de Gastos

```python
# Gráfico de dispersão
plt.scatter(df['Annual Income ($)'], df['Spending Score (1-100)'], c='blue', edgecolor='w', s=50)
plt.title('Renda Anual vs. Score de Gastos')
plt.xlabel('Renda Anual ($)')
plt.ylabel('Score de Gastos (1-100)')
plt.show()
```

```python
# Gráfico de dispersão
fig = px.scatter(df, x='Annual Income ($)', y='Spending Score (1-100)',
                 title='Renda Anual vs. Score de Gastos',
                 color='Gender',
                 hover_data=['Age', 'Profession'])
fig.update_layout(width=1000, height=600)
fig.show()
```

*  Score de Gastos por Faixa Etária

```python
# Criando faixas etárias
df['Faixa_Etaria'] = pd.cut(df['Age'], bins=[0, 18, 25, 35, 45, 55, 65, 100], labels=['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+'])
sns.boxplot(x='Faixa_Etaria', y='Spending Score (1-100)', data=df)
plt.title('Score de Gastos por Faixa Etária')
plt.show()
```

```python
# Criando faixas etárias
df['Faixa_Etaria'] = pd.cut(df['Age'], bins=[0, 18, 25, 35, 45, 55, 65, 100], labels=['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+'])

# Boxplot por faixa etária
fig = px.box(df, x='Faixa_Etaria', y='Spending Score (1-100)',
             title='Score de Gastos por Faixa Etária',
             color='Faixa_Etaria')
fig.update_layout(width=1000, height=600)
fig.show()
```
